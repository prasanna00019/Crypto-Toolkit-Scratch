```python
def convert_key_to_192bit(key):
    if len(key) != 24:
        raise ValueError("Key must be exactly 24 ASCII characters (192 bits).")
    # Convert the ASCII key to bytes and then to a hex string
    hex_key = key.encode('utf-8').hex()
    return hex_key.upper()  # Uppercase for readability
```

```python
def convert_input_to_128bit(input):
    n=len(input)
    if n != 16:
        input += '\x06'*((1+n//16)*16 - n)
    # Convert the ASCII input to bytes and then to a hex string
    hex_input = input.encode('utf-8').hex()
    return hex_input.upper()  
```

```python
def g_function(word, round_constant, sbox):
    rotated_word = word[1:] + word[:1]  
    #  SubWord - Apply S-Box Substitution
    subbed_word = [sbox[int(byte, 16)] for byte in rotated_word]  # Convert to int for lookup
    # XOR First Byte with Round Constant (Rcon[i])
    subbed_word[0] ^= round_constant
    # Convert to hexadecimal format
    return [format(byte, '02X') for byte in subbed_word]

def generate_all_keys(key, sbox, rcon):
    """
    Generates all 13 round keys for AES-192 (Key Expansion Algorithm)
    :param key: 24-character ASCII key
    :param sbox: AES S-Box table
    :param rcon: List of round constants
    :return: List of all 13 round keys (each 4x4 matrix)
    """
    # Convert key to 192-bit hex
    str_key = key.encode('utf-8').hex()
    # Initialize w[0] to w[5] from input key
    words = [[str_key[i:i+2] for i in range(j, j+8, 2)] for j in range(0, 48, 8)]  # 6 words (24 bytes)
    
    # Generate w[6] to w[51]
    for i in range(6, 52):
        if i % 6 == 0:
            g_word = g_function(words[i-1], rcon[i//6 - 1], sbox)
            new_word = [format(int(words[i-6][j], 16) ^ int(g_word[j], 16), '02X') for j in range(4)]
        else:
            new_word = [format(int(words[i-6][j], 16) ^ int(words[i-1][j], 16), '02X') for j in range(4)]
        words.append(new_word)
    
    # Group words into 13 round keys (each 4x4 matrix)
    round_keys = [words[i:i+4] for i in range(0, 52, 4)]
    return round_keys
```

```python
def transform_matrix(matrix):
    n = len(matrix)  # Get the size of the matrix
    return [[matrix[row][col] for row in range(n)] for col in range(n)]
def xor_hex_matrices(matrix1, matrix2):
    n = len(matrix1)  # Get matrix size
    # Compute XOR elementwise and format as uppercase hex strings
    result = [[format(int(matrix1[row][col], 16) ^ int(matrix2[row][col], 16), '02X')
               for col in range(n)] for row in range(n)]
    return result
def round_zero(input,key_R0):
    str_input= convert_input_to_128bit(input);
    input_mat= [[str_input[i:i+2] for i in range(j, j+8, 2)] for j in range(0, 32, 8)]
    transformed_input_mat=transform_matrix(input_mat);
    transformed_R0_key=transform_matrix(key_R0);
    state_matrix=xor_hex_matrices(transformed_input_mat,transformed_R0_key);
    return state_matrix
```

```python
def apply_sub_bytes(input_mat):
    return [[format(sbox[int(byte, 16)], '02X') for byte in row] for row in input_mat]
def shift_rows(input_mat):
    return [
        input_mat[0],  # Row 0 remains unchanged
        input_mat[1][1:] + input_mat[1][:1],  # Shift left by 1
        input_mat[2][2:] + input_mat[2][:2],  # Shift left by 2
        input_mat[3][3:] + input_mat[3][:3]   # Shift left by 3
    ]   
def galois_mult(a, b):
    result = 0
    for _ in range(8):
        if b & 1:
            result ^= a
        high_bit_set = a & 0x80
        a = (a << 1) & 0xFF
        if high_bit_set:
            a ^= 0x1B  # AES irreducible polynomial x^8 + x^4 + x^3 + x + 1
        b >>= 1
    return result
multiplying_matrix = [
    ['02', '03', '01', '01'],
    ['01', '02', '03', '01'],
    ['01', '01', '02', '03'],
    ['03', '01', '01', '02']
]
def mix_columns(input_mat, multiplying_mat):
    n = 4 
    output_mat = [[0] * n for _ in range(n)]
    
    for col in range(n):
        for row in range(n):
            output_mat[row][col] = 0
            for k in range(n):
                output_mat[row][col] ^= galois_mult(int(multiplying_mat[row][k], 16), int(input_mat[k][col], 16))
            output_mat[row][col] = format(output_mat[row][col], '02X')
    
    return output_mat
def add_round_key(mat1,mat2):
    return xor_hex_matrices(mat1,mat2);
```

```python
def Ith_Round(mat,i):
     output_sub_bytes_i=apply_sub_bytes(mat);
     output_shift_rows_i=shift_rows(output_sub_bytes_i);
     output_mix_columns_i=mix_columns(output_shift_rows_i,multiplying_matrix);   
     output_add_round_key_i=add_round_key(output_mix_columns_i,transform_matrix(round_keys[i+1]));
     return output_add_round_key_i
```

```python
def round_one_to_twelve(state_matrix):
    for i in range(11):
       if(i==0): 
        cip=Ith_Round(state_matrix,i)
       else:
        cip=Ith_Round(cip,i);  
    #round 12
    output_sub_bytes_12=apply_sub_bytes(cip);
    output_shift_rows_12=shift_rows(output_sub_bytes_12);
    cipher_text=add_round_key(output_shift_rows_12,transform_matrix(round_keys[12]));
    cipher=""
    cip_mat= transform_matrix(cipher_text) 
    for i in  range(len(cip_mat)):
        for j in range(len(cip_mat)):
            cipher+=cip_mat[i][j]+" ";
    return cipher        
```

```python
def inv_shift_rows(input_mat):
    return [
        input_mat[0],  # Row 0 remains unchanged
        input_mat[1][-1:] + input_mat[1][:-1],  # Shift right by 1
        input_mat[2][-2:] + input_mat[2][:-2],  # Shift right by 2
        input_mat[3][-3:] + input_mat[3][:-3]   # Shift right by 3
    ]

def inv_sub_bytes(input_mat):
    return [[format(sbox_inv[int(byte, 16)], '02X') for byte in row] for row in input_mat]

inv_multiplying_matrix = [
    ['0E', '0B', '0D', '09'],
    ['09', '0E', '0B', '0D'],
    ['0D', '09', '0E', '0B'],
    ['0B', '0D', '09', '0E']
]

def inv_mix_columns(input_mat, inv_multiplying_mat):
    n = 4  # AES block size
    output_mat = [[0] * n for _ in range(n)]
    
    for col in range(n):
        for row in range(n):
            output_mat[row][col] = 0
            for k in range(n):
                output_mat[row][col] ^= galois_mult(int(inv_multiplying_mat[row][k], 16), int(input_mat[k][col], 16))
            output_mat[row][col] = format(output_mat[row][col], '02X')
    
    return output_mat

def inv_round_zero(cipher_text, key_R12):
    cipher_mat = [[cipher_text[i:i+2] for i in range(j, j+8, 2)] for j in range(0, 32, 8)]
    transformed_cipher_mat = transform_matrix(cipher_mat)
    transformed_R12_key = transform_matrix(key_R12)
    state_matrix = xor_hex_matrices(transformed_cipher_mat, transformed_R12_key)
    return state_matrix

def inv_Ith_Round(mat, i):
    output_inv_shift_rows_i = inv_shift_rows(mat)
    output_inv_sub_bytes_i = inv_sub_bytes(output_inv_shift_rows_i)
    output_add_round_key_i = add_round_key(output_inv_sub_bytes_i, transform_matrix(round_keys[11 - i]))
    output_inv_mix_columns_i = inv_mix_columns(output_add_round_key_i, inv_multiplying_matrix)
    return output_inv_mix_columns_i

def inv_round_one_to_ten(state_matrix):
    for i in range(11):
        if i == 0:
            cip = inv_Ith_Round(state_matrix, i)
        else:
            cip = inv_Ith_Round(cip, i)
    
    # Round 10
    output_inv_shift_rows_12 = inv_shift_rows(cip)
    output_inv_sub_bytes_12 = inv_sub_bytes(output_inv_shift_rows_12)
    plain_text = add_round_key(output_inv_sub_bytes_12, transform_matrix(round_keys[0]))
    plain = ""
    plain_mat = transform_matrix(plain_text)
    for i in range(len(plain_mat)):
        for j in range(len(plain_mat)):
            plain += plain_mat[i][j] + " "
    return plain
```

```python
def AES_Decrypt(cipher_text, key):
    cipher_text_clean = cipher_text.replace(" ", "")
    round_keys = generate_all_keys(key, sbox, rcon)
    state_matrix = inv_round_zero(cipher_text_clean, round_keys[12])
    decrypted_hex = inv_round_one_to_ten(state_matrix)
    return decrypted_hex
```

```python
def hex_to_plaintext(hex_string):
    try:
        # Convert hex string to bytes
        byte_data = bytes.fromhex(hex_string)
        # Decode bytes to plaintext using ASCII encoding
        plaintext = byte_data.decode('ascii')
        return plaintext
    except ValueError as e:
        # Handle invalid hex strings or decoding errors
        return f"Error: {e}"
```

```python
def hex_to_string(hex_input):
    utf8_bytes = bytes.fromhex(hex_input)  
    text = utf8_bytes.decode('utf-8')     
    return text
```
